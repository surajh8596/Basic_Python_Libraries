```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()

axis=fig.add_axes([0.5,0.1,0.8,0.8]) # control the left,right,width,height of the canva(from 0-1)

axis.plot(x,y,'r')
```

```python
# subplot
fig,axes=plt.subplots(nrows=1,ncols=2)  # submap is of 1 row , 2 columns
for ax in axes:
    ax.plot(x,y,'g')
```

```python
# subplot
fig,axes=plt.subplots(nrows=2,ncols=1)  # submap is of 1 row , 2 columns
for ax in axes:
    ax.plot(x,y,'g')
```

```python
# we can also plot graph inside another graph
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) # big axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # small axes

axes1.plot(x,y,'r')
axes2.plot(x,y,'g');
```

```python
# dpi - dots per inches(quality of graph, more the dpi more the memory use)

fig=plt.figure(figsize=(6,6), dpi=50) 

fig.add_subplot()
plt.plot(x,y,'b--');
```

```python
# legend, title , label, line color, transparency

ax.legend(["label-1","label-2"])
fig, axes=plt.subplots(dpi=200)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title of graph")

axes.plot(x,x**2, color='blue', alpha=0.5)
axes.plot(x,x**3, color='red', alpha=0.1)

axes.legend(["y=x**2","y=x**3"],loc=0);
```

```python
# legend, title , label, line color, transparency, marker

ax.legend(["label-1","label-2","label-3"])
fig, axes=plt.subplots(dpi=200)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title of graph")

axes.plot(x,x+1, color='blue', alpha=0.5, lw=2, linestyle='-',marker='o')
axes.plot(x,x+2, color='red', alpha=0.4, lw=2, linestyle='-', marker='s')
axes.plot(x,x+3, color='black', alpha=0.3, lw=2, linestyle='-',marker='*', markersize=20, markerfacecolor='green')


axes.legend(["y=x+1","y=x+2","y=x+3"],loc=0);
```

```python
# limit the x and y axis

fig,axes=plt.subplots(1,2,figsize=(10,5))

axes[0].plot(x,x**2, x, x**3, lw=2)
axes[0].grid(True)

axes[1].plot(x,x**2, x, x**3, lw=2)
axes[1].grid(True)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5]);
```

## 2D graph

```python
n=np.array([0,1,2,3,4,5])
fig, axes=plt.subplots(1,4, figsize=(16,5))
axes[0].set_title("Scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("Step")
axes[1].step(n,n**2, lw=2)

axes[2].set_title("Bar")
axes[2].bar(n,n**2, align='center', width=0.5,alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(n,n**2,n**3, color='green', alpha=0.5)
```

```python
# radar chart

fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,1,1],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*2,color='red',lw=2);
```

```python
# histogram

n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulutive detained histogram")
axes[1].hist(n, cumulative=True, bins=5);
```

#### contour map

```python
import matplotlib.cm as cm  ## COLORMAP(CM)
```

```python
delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2

print(X)
print(Y)
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title("CONTOUR MAP");
```
