```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
from scipy.cluster.vq import kmeans, whiten
```

```python
sns.set_style("darkgrid")
ptc=100
a=np.random.multivariate_normal([0,0],[[4,1],[1,4]], size=ptc)
b=np.random.multivariate_normal([30,10],[[10,2],[2,1]], size=ptc)
features=np.concatenate((a,b))
whitened=whiten(features)
codebook, distortion=kmeans(whitened, 2)

plt.scatter(whitened[:,0],whitened[:,1])
plt.scatter(codebook[:,0],codebook[:,1], c='r')
plt.show()
```

## FFTpack

```python
fre=10
fre_samp=100
t=np.linspace(0,2,2*fre_samp, endpoint=False)
a=np.sin(fre*2*np.pi*t)
plt.plot(t,a)
plt.xlabel('Time in (s)')
plt.ylabel('Signal aptitude')
plt.show()
```

```python
from scipy import fftpack
```

```python
A=fftpack.fft(a)
frequency=fftpack.fftfreq(len(a))*fre_samp
plt.stem(frequency, np.abs(A),use_line_collection=True)
plt.xlabel('Frequency in Hz')
plt.ylabel('Frequency spectrum mangitude')
plt.show()
```
